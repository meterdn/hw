```python
import random
class game():
    def __init__(self):
        self.swords = 0
        self.potions = 0
        self.alive = True
        self.passed = []
        self.point = 0
     #   self.konumx = 0
     #   self.konumy = 0
        self.konum = ()
    #def update(self):
    #    self.konum = (self.konumx,self.konumy)
    def map_generator(self):
        potions = []
        swords = []
        venoms = []
        monsters = []
        treasures = []
        mapp = []
        determined = []
    
        for i in range(0,6):
            mapp.append([])
            for j in range(0,7):
                mapp[i].append(" ")
            
        while(len(treasures)<5):
            x = random.randint(0,5)
            y = random.randint(0,6)
            if (x,y) not in determined:
                treasures.append((x,y))
                determined.append((x,y))
        while(len(venoms)<3):
            x = random.randint(0,5)
            y = random.randint(0,6)
            if (x,y) not in determined:
                venoms.append((x,y))
                determined.append((x,y))
        while(len(potions)<3):
            x = random.randint(0,5)
            y = random.randint(0,6)
            if (x,y) not in determined:
                potions.append((x,y))
                determined.append((x,y))
        while(len(monsters)<5):
            x = random.randint(0,5)
            y = random.randint(0,6)
            if (x,y) not in determined:
                monsters.append((x,y))
                determined.append((x,y))
        while(len(swords)<2):
            x = random.randint(0,5)
            y = random.randint(0,6)
            if (x,y) not in determined:
                swords.append((x,y))
                determined.append((x,y))
        x = random.randint(0,5)
        y = random.randint(0,6)
        while (x,y) in determined:
            x = random.randint(0,5)
            y = random.randint(0,6)
        mapp[x][y] = "E"
        self.passed.append((x,y))
        self.konum = (x,y)
        print(mapp)
        while(self.alive):
            move = input("press L or l, R or r, U or u, D or d to move:")
            if move == "L" or move == "l":
                if self.konum[1]>0 and (self.konum[0],self.konum[1]-1) not in self.passed:
                    self.konum = (self.konum[0],self.konum[1]-1)
                    self.point += 1
                    self.passed.append(self.konum)
            elif move == "R" or move == "r":
                if self.konum[1]<6 and (self.konum[0],self.konum[1]+1) not in self.passed:
                    self.point += 1
                    self.konum = (self.konum[0],self.konum[1]+1)
                    self.passed.append(self.konum)
            elif move == "U" or move == "u":
                if self.konum[0]>0 and (self.konum[0]-1,self.konum[1]) not in self.passed:
                    self.point += 1
                    self.konum = (self.konum[0]-1,self.konum[1])
                    self.passed.append(self.konum)
            elif move == "D" or move == "d":
                if self.konum[0]<5 and (self.konum[0]+1,self.konum[1]) not in self.passed:
                    self.point += 1
                    self.konum = (self.konum[0]+1,self.konum[1])
                    self.passed.append(self.konum)
            if self.konum in swords:
                self.swords += 1
                mapp[self.konum[0]][self.konum[1]] = "S"
            elif self.konum in potions:
                self.potions += 1
                mapp[self.konum[0]][self.konum[1]] = "P"
            elif self.konum in venoms:
                if self.potions >0:
                    self.potions -= 1
                    mapp[self.konum[0]][self.konum[1]] = "V"
                else:
                    mapp[self.konum[0]][self.konum[1]] = "V"
                    self.alive = False
                    break
            elif self.konum in monsters:
                if self.swords >0:
                    mapp[self.konum[0]][self.konum[1]] = "M"
                    self.swords -= 1
                else:
                    mapp[self.konum[0]][self.konum[1]] = "M"
                    self.alive = False
                    break
            else:
                mapp[self.konum[0]][self.konum[1]] = "E"
        print("----------------------------\n")
        for i in range(0,len(mapp)):
            print(mapp[i])
            print("\n")
        print("----------------------------\n")
        print("You die. \n Score:{} Sword: {} Potion: {}".format(self.point,self.swords,self.potions))
obj = game()
obj.map_generator()
```
